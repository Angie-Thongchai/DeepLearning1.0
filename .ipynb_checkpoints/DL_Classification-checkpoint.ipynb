## Processing Data

- Standardize the data by scaling the image

````python
# use the lamda function to scaling images
normal=raw_data.map(lambda x, y: (x/255.0, y))

# convert value to numpy array
batch_ds=next(normal.as_numpy_iterator())

# present values and the results 
print(f"Minimun value: {batch_ds[0].min()}, Maximun value: {batch_ds[0].max()}")
print(f'Data is split into {len(normal)} batches and Each batch has{len(batch_ds[0])} images')
print(f"A sample of batchs: {batch_ds[:1]}")
````

## split the data

- Split the data for a training, a validation and a test.
- Therefore, the tolal number of a split size is 23.

````python
# based on batch's value is 23 batches. 

train_size = int(len(normal)* 0.7)
val_size = int(len(normal)* 0.2) + 1
test_size = int(len(normal)* 0.1) 
````

````python
# Show the details

print(f"A train's size:{train_size}, A validation's size:{val_size}, A test's size:{test_size}")
print(f"the total value: {train_size+ val_size+ test_size}" )
````

````python
# Split the data and use the data size

train = normal.take(train_size)
val = normal.skip(train_size).take(val_size)
test = normal.skip(train_size + val_size).take(test_size)
````

## Create the model and compile

* The stride can reduce the resolution of the output

````python
with strategy.scope():
    model = Sequential()

    model.add(Conv2D(filters= 16, kernel_size= (3,3), strides= 1, activation= 'relu', input_shape= (256, 256, 3)))
    model.add(MaxPooling2D())

    model.add(Conv2D(filters= 32, kernel_size= (3,3), strides= 1, activation= 'relu'))
    model.add(MaxPooling2D())

    model.add(Conv2D(filters= 16, kernel_size= (3,3), strides= 1, activation= 'relu'))
    model.add(MaxPooling2D())

    model.add(Flatten())

    model.add(Dense(units= 256, activation= 'relu'))
    model.add(Dense(units= 11, activation= 'softmax'))
    
    model.compile(optimizer= 'adam', loss= tf.losses.sparse_categorical_crossentropy, metrics = ['accuracy'] )
````

## Summary the result

````python
model.summary()
````

- TensorBoad's callback log for:

    * Metrics summary plot.
    
    * Trianing grap visualization.
    
    * Weight histograms.

````python
# Use tensorBoard's callback.

tensorboard = tf.keras.callbacks.TensorBoard(log_dir = 'logs')
````

````python
%%time
# Train the model

with strategy.scope():
    hist = model.fit(train, epochs= 10, validation_data= val, callbacks= [tensorboard])
````

##  Model Evaluation

````python
# Use visualization to evaluate the model.

acc=hist.history['accuracy']
val_acc=hist.history['val_accuracy']

loss=hist.history['loss']
val_loss=hist.history['val_loss']

epochs_range=range(10)

plt.figure(figsize=(8, 8))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()
````

- CategloricalAccuracy to merge the state from one or more metrics (Typically the state will be stored in the form of the metric's weights).

````python
# Use CategloricalAccuracy to evaluate the model

acc = CategoricalAccuracy()

for batch_test in test.as_numpy_iterator():
    x,y = batch_test
    yhat = model.predict(x)
    yhat = [i.argmax() for i in yhat]
    acc.update_state(y, yhat)
````

## Data augmentation

it is a technique of artificially increasing the training set by creating modified copies of a dataset using existing data. It includes making minor changes to the dataset or using deep learning to generate new data points. 

````python
# create an augmentation by use horizontal and size=256*256

data_augmentation = keras.Sequential(
  [
    layers.RandomFlip("horizontal",input_shape=(256,256,3)),
    layers.RandomRotation(0.1),
    layers.RandomZoom(0.1),
  ]
)
````

````python
# Visualize an example of images by using data augmentation

plt.figure(figsize=(10, 10))
for images,labels in raw_data.take(1):
  for i in range(9):
    augmented_images = data_augmentation(images)
    ax = plt.subplot(3, 3, i + 1)
    plt.imshow(augmented_images[0].numpy().astype(int))
    plt.axis("off")
````

## Dropout the dataset and use data augmentation

````python
with strategy.scope():
    model3 = Sequential(data_augmentation)

    model3.add(Conv2D(filters= 16, kernel_size= (3,3), strides= 1, activation= 'relu', input_shape= (256, 256, 3)))
    model3.add(MaxPooling2D())

    model3.add(Conv2D(filters= 32, kernel_size= (3,3), strides= 1, activation= 'relu'))
    model3.add(MaxPooling2D())

    model3.add(Conv2D(filters= 16, kernel_size= (3,3), strides= 1, activation= 'relu'))
    model3.add(MaxPooling2D())

    model3.add(Flatten())

    model3.add(Dense(units= 256, activation= 'relu'))
    model3.add(Dense(units= 11, activation= 'softmax'))
    
    model3.compile(optimizer= 'adam', loss= tf.losses.sparse_categorical_crossentropy, metrics = ['accuracy'] )
````

````python
model3.summary()
````

## Training the model with data augmentation

````python
%%time
# Train the model3 epochs=30

with strategy.scope():
    hist3 = model3.fit(train, epochs= 30, validation_data= val, callbacks= [tensorboard])
````

## Model Evaluation with data augmentation

````python
# Use visualization to evaluate the model.

acc=hist3.history['accuracy']
val_acc=hist3.history['val_accuracy']

loss=hist3.history['loss']
val_loss=hist3.history['val_loss']

epochs_range=range(30)

plt.figure(figsize=(8, 8))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()
````
